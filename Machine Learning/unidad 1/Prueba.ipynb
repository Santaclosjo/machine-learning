```python
import pandas as pd
```

```python
DF=pd.read_excel("isl-012021.xlsx" ,index_col="Región")
DF.head()
```



Filtramos datos que no contengan nada





```python

DFF=DF[DF["Monto m$"]>0]
DFF.groupby("Glosa Comuna").agg({
  "Glosa Comuna":'sum',
  "Nº Mujer":'sum',
  " Mto.Mujer":'sum',
  " Nº Hombre":'sum',
  " Mto.Hombre":'sum'
  })

DN=DFF.groupby("Glosa Comuna").agg({
  "Glosa Comuna":'sum',
  "Nº Mujer":'sum',
  " Mto.Mujer":'sum',
  " Nº Hombre":'sum',
  " Mto.Hombre":'sum'
})

DN.describe()
```
